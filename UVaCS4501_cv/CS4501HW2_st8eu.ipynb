**1.1 ** Compute and display $I_x^2$, $I_y^2$, and $I_xI_y$ using Sobel operators. You are only allowed to use scipy.signal.convolve2d for any convolutional operations. Feel free to reuse your code from the previous assignment. (1pts)

```python
from scipy import signal

# Compute Ixx, Iyy and Ixy here.
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])


img_gray = (img[:, :, 0] + img[:, :, 1] + img[:, :, 2]) / 3.0
Ix = signal.convolve2d(img_gray, sobel_x, mode='same')
Iy = signal.convolve2d(img_gray, sobel_x.T, mode='same')

Ixx = Ix
Iyy = Iy
Ixy = Ix * Iy

# Plot Ixx, Iyy, Ixy.
plt.figure(figsize=(20,4))
plt.imshow(np.concatenate((Ixx, Iyy, np.abs(Ixy)), axis = 1), 
           cmap =  'jet')
plt.colorbar()
plt.grid('off')
plt.show()
```

[链接文字](https://)**1.2** Show side by side the input image with the response map R computed using the original Harris detector with $\alpha=0.05$, and using the Harmonic mean formula proposed by Szeliski. Use a window/patch size of 12x12 to define the momentum matrix M. (2pts).

```python
# Compute the Rmap and RHarmonicMap map responses.
import math

def gussian(ks=12, sigma=2):
  gussian_kernal = np.ones((ks,ks))
  for x in range(ks):
    for y in range(ks):
      G = 1/(2*math.pi*(sigma**2)) * math.exp(-((x-ks//2)**2 + (y-ks//2)**2)/(2*sigma**2))
      gussian_kernal[x,y] = G
  return gussian_kernal

Gxx = signal.convolve2d(Ixx, gussian(), mode='same')
Gyy = signal.convolve2d(Iyy, gussian(), mode='same')
Gxy = signal.convolve2d(Ixy, gussian(), mode='same')

M = [[Gxx, Gxy],
     [Gxy, Gyy]]

alpha = 0.05
det = Gxx * Gyy - np.square(Gxy)
trace = Gxx + Gyy
RMap = det - alpha * np.square(trace)
RHarmonicMap = det / trace

# Plot Harris response outputs.
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey = True, figsize=(20, 4))
ax1.imshow(img);ax1.grid('off')
ax2.imshow(RMap, cmap = 'jet'); ax2.grid('off')
ax3.imshow(RHarmonicMap, cmap ='jet'); ax3.grid('off');
```

**1.3** Perform non-max supression on the outputs obtained in the previous step so that you get a list of pixels corresponding to corners for each method. Feel free to use any  method you can think of but the general idea is to leave only one point per corner and no two points with high values should be together. An easy way is to just define a minimum distance between corners of 5 pixels. The only pixels surviving (non-zero) after non-max should be the ones that a) pass a threshold (which you are free to define), and b) are the maximum pixel within their neighborhood. I am including outputs as an example of what it should more or less look like at the end. (2pts).

```python
def nonmax(rmap):
  # Non-max supression and thresholding function.
  threshold = 0.35;
  h, w = rmap.shape
  dx = 12; dy = 12;
  for x in range(h-dx+1):
    for y in range(w-dy+1):
      window = rmap[x:x+dx, y:y+dy]
      max = window.max()
      window[(window < max) | (window < threshold)] = 0
  return rmap

# np.nonzero extracts 2d coordinates of non-zero elements.
ys1, xs1 = np.nonzero(nonmax(RMap))
ys2, xs2 = np.nonzero(nonmax(RHarmonicMap))

# Method 1: Harris operator original score.
# Show image with corner coordinates marked with blue dots.
fig, ax = plt.subplots()
ax.imshow(img, interpolation='nearest', cmap=plt.cm.gray)
ax.plot(xs1, ys1, '.b', markersize=12)
ax.grid('off')

# Method 2: Szeliski's Harmonic mean.
# Show image with corner coordinates marked with blue dots.
fig, ax = plt.subplots()
ax.imshow(img, interpolation='nearest', cmap=plt.cm.gray)
ax.plot(xs2, ys2, '.b', markersize=12)
ax.grid('off')
```

### 3. Hough Transform for line detection (5pts).
In this section we will perform line detection by implementing the Hough Transform.

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen
from scipy import signal

# Try maybe downloading the images in case the CS webserver goes down.
img_url = 'http://www.cs.virginia.edu/~vicente/vision/uvacs.jpg'
img = sp.misc.imread(urlopen(img_url), mode = 'RGB')
img = img.astype('float') / 255

# Show the image size.
print('Image size: ', img.shape)

# Plot the image.
def display_image(img):
  plt.figure()
  plt.imshow(img)
  plt.grid('off')
  plt.show()
  
display_image(img)
```

**3.1** Use the Hough Transform to find all the lines in the image. Show the Hough Transform for the Rice Hall image (2pts).

```python
# find and thershold the magnitude map of the image (reusing codes from last assignment)
sobel_mask = np.array([[-1,0,1],
                       [-2,0,2],                       
                       [-1,0,1]])
img_grey = (img[:, :, 0] + img[:, :, 1] + img[:, :, 2]) / 3.0
magmap_x = signal.convolve2d(img_grey, sobel_mask, 'same', boundary='symm')
magmap_y = signal.convolve2d(img_grey, sobel_mask.T, 'same', boundary='symm')
magnitude_map = np.sqrt(magmap_x**2 + magmap_y**2)
magnitude_map[magnitude_map < 2.5] = 0

#extract feature points from magnitude_map
h, w, _ = img.shape
rou_max = int(np.sqrt(h**2 + w**2))
ys, xs = np.nonzero(magnitude_map)
hough_space = np.zeros((181, rou_max))


for x, y in zip(xs,ys):
  for theta in range(181):
    costheta = np.cos(np.deg2rad(theta))
    sintheta = np.sin(np.deg2rad(theta))
    rou = int(x * costheta + y * sintheta)
    hough_space[theta][rou] += 1
    
display_image(hough_space)
```

**3.2** Find the peaks in the Hough Transform computed in the previous step and plot the original images with red lines overlaid showing the detected lines (3pts).

```python
import copy
#computing hough_space is computationally expensive so I kept a copy of it.
hs = copy.deepcopy(hough_space)

def nonmax(input):
  # Non-max supression and thresholding function.
  threshold = 38
  h, w = input.shape
  dx = 10; dy = 10;
  for x in range(h-dx+1):
    for y in range(w-dy+1):
      window = input[x:x+dx, y:y+dy]
      max = window.max()
      window[(window < max)| (window < threshold)] = 0
      input[x:x+dx, y:y+dy] = window
  return input

hs = nonmax(hs)
thetas, rous = np.nonzero(hs)

fig, ax = plt.subplots()
ax.imshow(img, interpolation='nearest', cmap=plt.cm.gray)

for theta, rou in zip(thetas,rous):
  costheta = np.cos(np.deg2rad(theta))
  sintheta = np.sin(np.deg2rad(theta))
  m = -costheta/sintheta
  b = rou / sintheta
  
  x = np.linspace(0, w)
  ax.plot(x, m*x+b, 'r')
  ax.set_ylim(h, 0)
  
ax.grid('off')
```
